```python
import os 
import mmcv
```

```python
root_path = '../data/balloon_dataset/balloon/train/'
anno_file = os.path.join(root_path, 'via_region_data.json')
output_file = os.path.join(root_path, 'coco_via_region_data.json')

data_infos = mmcv.load(ann_file)
data_infos[list(data_infos.keys())[0]]
```

```python
def convert_annofile_to_coco(anno_file, output_file, root_path) :
    data_infos = mmcv.load(anno_file)
    
    annotations = []
    images = []
    obj_cnt = 0
    for idx, v in enumerate(mmcv.track_iter_progress(data_infos.values())) :
        file_path = os.path.join(root_path, v['filename'])

        # image
        width, height = mmcv.imread(file_path).shape[:2]
        file_name = v['filename']

        images.append({
            'id' : idx,
            'width' : width,
            'height' : height,
            'file_name' : file_name
        })

        # annoitation
        for k, obj in v['regions'].items() :
            assert not obj['region_attributes']
            px = obj['shape_attributes']['all_points_x']
            py = obj['shape_attributes']['all_points_y']                
            poly = [(x + 0.5, y + 0.5)for x, y in zip(px, py)]
            poly = [p for x in poly for p in x]
            x = min(px)
            y = min(py)
            width = max(px) - min(px)
            height = max(py) - min(py)

            annotations.append({
                'id' : obj_cnt,
                'image_id' : idx,
                'category_id' : 0,
                'segmentation' : [poly],
                'bbox' : [x,y,width, height],
                'area' : width * height,
                'iscrowd' : 0
            })

            obj_cnt += 1

    # categories
    categories_info = {
        'id' : 0,
        'name' : 'balloon'
    }



    coco_format = dict(
        images = images,
        annotations = annotations,
        categories = [{
            'id' : 0,
            'name' : 'balloon'
        }]
    )

    mmcv.dump(coco_format, output_file)
```

```python
root_path = '../data/balloon_dataset/balloon/train/'
anno_file = os.path.join(root_path, 'via_region_data.json')
output_file = os.path.join(root_path, 'coco_via_region_data.json')
convert_annofile_to_coco(anno_file, output_file, root_path)

root_path = '../data/balloon_dataset/balloon/val/'
anno_file = os.path.join(root_path, 'via_region_data.json')
output_file = os.path.join(root_path, 'coco_via_region_data.json')
convert_annofile_to_coco(anno_file, output_file, root_path)
```
